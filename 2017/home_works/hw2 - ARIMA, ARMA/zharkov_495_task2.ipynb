```python
from copy import deepcopy
my_ts = pd.DataFrame(ts_dep42['item 52425'])
my_ts.dropna(inplace=True)
print(my_ts.shape)
train_end_ind = int(my_ts.shape[0] * .8)
train_ts = deepcopy(my_ts.iloc[:train_end_ind])
test_ts = my_ts.iloc[train_end_ind:]
train_ts.plot(figsize=(15,5))
```

```python
from statsmodels.tsa import stattools
stattools.kpss(np.array(train_ts).reshape(-1,))
```

### d=0

```python
import statsmodels.api as sm

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(train_ts.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(train_ts.values.squeeze(), lags=40, ax=ax2)
```

```python
q_candidates = list(range(8))
p_candidates = [0,1,3,4,6,7]
d = 0
```

```python
%%time
import warnings
warnings.filterwarnings('ignore')
for p in p_candidates:
    for q in q_candidates:
        print('p={}\td={}\tq={}'.format(p,d,q))
        try:
            model = sm.tsa.ARMA(train_ts.values.squeeze(), (p, q)).fit()
            print(model.aic, model.bic, model.hqic)
        except:
            print('failed')
```

```python
from scipy import stats
import pylab
best_p, best_q = 6, 3
model = sm.tsa.ARMA(train_ts, (best_p, best_q)).fit()
# check unbiaseness
print('durbin-watson-stat:', sm.stats.durbin_watson(model.resid.values))
stat = stats.wilcoxon(model.resid.values)
print('Statistic: %f\npvalue: %.2f' %(stat[0], stat[1]))
# stationarity test
print()
print('stationarity:')
print(stattools.kpss(model.resid))
# autocorrelation test
print()
print('autocorrelation:')
print(stats.ttest_1samp(model.resid, 0))
from utils import qualityMACAPE
## q-qPlot
# fig = plt.figure()
# res = stats.probplot(model.resid, dist="norm", plot=pylab)
```

### all tests passed well

```python
from tqdm import tqdm
predictions = []
history = my_ts
for t in tqdm(range(len(test_ts)), position=0):
    model = sm.tsa.ARMA(history.iloc[:len(train_ts) + t], order=(best_p, best_q))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
```

```python
test_ts['forecast'] = np.array(predictions).reshape(-1,)
```

```python
print('MACAPE:', qualityMACAPE(test_ts[test_ts.columns[0]], test_ts['forecast'])[0])
test_ts.plot(figsize=(15,5))
```
